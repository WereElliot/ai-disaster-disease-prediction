# Phase 2: Exploratory Data Analysis - Visualizations
Time-series plots, correlation heatmaps, and spatial distribution mapping.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import os
import sys

# Ensure project root is in path
sys.path.append(os.path.abspath('..'))
from src.utils.config import load_config

config = load_config('../config.yaml')
processed_path = Path('..') / config['data']['paths']['processed'] / 'final_dataset.parquet'
df = pd.read_parquet(processed_path)

# Ensure output directories exist
plots_dir = Path('../docs/plots')
plots_dir.mkdir(parents=True, exist_ok=True)

plt.style.use('ggplot')
df.head()
```

## 1. Time-Series Analysis

```python
plt.figure(figsize=(15, 6))
df.set_index('date')['cases'].plot(title='Disease Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.savefig(plots_dir / 'cases_over_time.png')
plt.show()
```

## 2. Correlation Analysis

```python
plt.figure(figsize=(10, 8))
corr = df.drop(columns=['date', 'location']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.savefig(plots_dir / 'correlation_heatmap.png')
plt.show()
```

## 3. Spatial Distribution (Nairobi Only for Sample)

```python
location_summary = df.groupby('location')['cases'].sum().reset_index()
sns.barplot(data=location_summary, x='location', y='cases')
plt.title('Cases by Location')
plt.savefig(plots_dir / 'cases_by_location.png')
plt.show()
```
