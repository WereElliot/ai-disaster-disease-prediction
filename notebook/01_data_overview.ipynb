```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

rows = 500

data = {
    "date": pd.date_range(start="2020-01-01", periods=rows, freq="D"),
    "temperature": np.random.normal(25, 5, rows),
    "humidity": np.random.uniform(40, 90, rows),
    "rainfall": np.random.exponential(5, rows),
    "population_density": np.random.randint(50, 1000, rows),
    "flood_event": np.random.choice([0,1], rows),
    "disease_cases": np.random.poisson(10, rows)
}

df = pd.DataFrame(data)

df.head()
```

```python
df.to_csv("../data/sample_disease_climate_data.csv", index=False)
```
